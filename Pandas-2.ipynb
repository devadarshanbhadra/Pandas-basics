#### 5. Applying a function which seperated fare pay into groups

```python
# seperating into categories based on fare
```

* 0 - 100 -  'A'
* 101- 199 - 'B'
* 200+     - 'C' 

```python
def test3(x):
    if x<100:
        return "A"
    elif x>100 and x<200:
        return "B"
    else:
        return "C"
```

```python
data['Fare_group'] = data['Fare'].apply(test3)
```
